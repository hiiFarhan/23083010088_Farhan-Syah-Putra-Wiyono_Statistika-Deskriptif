### Visualisasi Data ke 1. Taxable Amount

```python
# Mengatur tampilan global untuk memastikan tidak ada notasi 'e' dan hanya menampilkan 2 angka di belakang koma
pd.options.display.float_format = '{:.2f}'.format

# Memilih kolom yang relevan untuk analisis
Taxable_Amount = df['Taxable Amount']

# Membuat plot data
plt.figure(figsize=(8, 4))
sns.histplot(df['Taxable Amount'], kde=True, bins=30, color='#00008B')

plt.title('Distribusi dari Taxable Amount')
plt.xlabel('Taxable Amount')
plt.ylabel('Frekuensi')
plt.show()

def describe_manual(data, columns):
    results = []
    for col in columns:
        mean = data[col].mean()  # Menghitung rata-rata
        mode = data[col].mode()[0]  # Menghitung modus (nilai yang paling sering muncul)
        median = data[col].median()  # Menghitung median (nilai tengah)
        variance = data[col].var()  # Menghitung varians
        std_dev = data[col].std()  # Menghitung standar deviasi
        q1 = data[col].quantile(0.25)  # Menghitung quartile 1 (25%)
        q2 = data[col].quantile(0.50)  # Menghitung quartile 2 (50%, atau median)
        q3 = data[col].quantile(0.75)  # Menghitung quartile 3 (75%)
        range_val = data[col].max() - data[col].min()  # Menghitung rentang nilai (maks - min)
        
        # Menambahkan hasil ke dalam dictionary
        results.append({
            'Mean': round(mean, 2),
            'Median': round(median, 2),
            'Mode': round(mode, 2),
            'Range': round(range_val, 2),
            'Standar Deviation': round(std_dev, 2),
            'Variance': round(variance, 2),
            'Quartile 1': round(q1, 2),
            'Quartile 2': round(q2, 2),
            'Quartile 3': round(q3, 2),
            'Skewness': round(data[col].skew(), 2)  # Menghitung skewness
        })

    # Mengubah hasil ke dalam bentuk DataFrame
    df_results = pd.DataFrame(results, index=columns)
    
    # Transpose DataFrame agar sesuai dengan format yang diinginkan
    df_results = df_results.transpose()

    # Mengatur nama kolom agar sesuai dengan tampilan yang diinginkan
    df_results.columns = ['Taxable Amount']

    return df_results

# Menerapkan fungsi pada kolom 'Taxable Amount'
stats = describe_manual(df, ['Taxable Amount'])

# Menampilkan hasilnya
stats
```

```python
print("\nAnalisis Skewness dari Taxable Amount:")
print("Data ini termasuk ke dalam distribusi yang skewness positif.")
print("Alasan: Skewness: 2.74 (positif), menunjukkan distribusi yang miring ke kanan.")
print("Ini berarti mayoritas nilai pajak hotel cenderung rendah,")
print("namun terdapat beberapa perusahaan dengan jumlah pajak kena yang jauh lebih tinggi dibandingkan yang lain,")
print("sehingga menciptakan ekor distribusi di sisi kanan.")
```
